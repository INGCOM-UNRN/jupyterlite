## 6. Banderas de Control

### Ejercicio Resuelto: Buscar n√∫mero par

**Problema:** Buscar el primer n√∫mero par en una secuencia.

**Soluci√≥n paso a paso:**

````python
# Paso 1: Inicializar bandera y contador
encontrado = False
numero = 1

# Paso 2: Buscar mientras no encontremos
while not encontrado and numero <= 10:
    if numero % 2 == 0:
        print(f"Primer par encontrado: {numero}")
        encontrado = True  # Cambiar bandera
    numero = numero + 1

# Paso 3: Verificar resultado
if not encontrado:
    print("No se encontr√≥ ning√∫n par")

# Explicaci√≥n:
# - Bandera: variable booleana para controlar el loop
# - Inicializar en False
# - Cambiar a True cuando se cumple la condici√≥n
# - El loop termina cuando la bandera es True
````

## 8. Control de Bucles: `break` y `continue`

### Ejercicio Resuelto: Buscar con break

**Problema:** Buscar la primera vocal en un texto.

**Soluci√≥n paso a paso:**

````python
# Paso 1: Texto de prueba
texto = "Python es genial"

# Paso 2: Buscar con break
encontrada = False
for letra in texto:
    if letra in "aeiouAEIOU":
        print(f"Primera vocal encontrada: {letra}")
        encontrada = True
        break  # Salir del loop inmediatamente

if not encontrada:
    print("No hay vocales")

# Paso 3: Ejemplo de continue (saltar iteraci√≥n)
print("\nN√∫meros impares del 1 al 10:")
for i in range(1, 11):
    if i % 2 == 0:  # Si es par
        continue    # Saltar esta iteraci√≥n
    print(i)  # Solo se ejecuta para impares

# Explicaci√≥n:
# - break: sale del loop completamente
# - continue: salta a la siguiente iteraci√≥n
````

## 9. Bucles Anidados

### Ejercicio Resuelto: Tabla de multiplicar completa

**Problema:** Mostrar tablas del 1 al 3.

**Soluci√≥n paso a paso:**

````python
# Paso 1: Bucle externo (tabla)
for tabla in range(1, 4):  # Tablas del 1 al 3
    print(f"\n=== Tabla del {tabla} ===")
    
    # Paso 2: Bucle interno (multiplicador)
    for multiplicador in range(1, 6):  # Del 1 al 5
        resultado = tabla * multiplicador
        print(f"{tabla} x {multiplicador} = {resultado}")

# Explicaci√≥n:
# - El bucle externo controla qu√© tabla
# - El bucle interno hace las multiplicaciones
# - Por cada iteraci√≥n del externo, el interno se ejecuta completamente
# - Total iteraciones: 3 √ó 5 = 15
````

## 10. Lectura de C√≥digo

En esta secci√≥n, **NO ejecutes el c√≥digo inmediatamente**. Primero le√© y determin√° qu√© hace.

### Instrucciones:
1. Le√© el c√≥digo cuidadosamente
2. Intent√° predecir la salida en un comentario
3. Reci√©n despu√©s ejecut√° para verificar

### C√≥digo 10.1: ¬øQu√© imprime? (F√°cil)

````python
# ¬øQu√© imprime este c√≥digo?
# Tu predicci√≥n aqu√≠ antes de ejecutar:

x = 10
if x > 5:
    print("Grande")
if x > 15:
    print("Muy grande")
print("Fin")

# Ejecut√° y verific√° tu predicci√≥n
````

### C√≥digo 10.2: ¬øQu√© hace este while? (F√°cil-Medio)

````python
# ¬øQu√© hace este c√≥digo? Describ√≠ en un comentario:

contador = 5
suma = 0
while contador > 0:
    suma = suma + contador
    contador = contador - 1
print(suma)

# Tu respuesta: Este c√≥digo...
# Ejecut√° y verific√°
````

### C√≥digo 10.3: ¬øCu√°ntas veces se ejecuta? (Medio)
