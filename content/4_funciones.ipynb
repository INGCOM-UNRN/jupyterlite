### Ejercicio 6.3: Docstring con ejemplos

**Problema:** Agreg√° ejemplos de uso en el docstring:

````python
def factorial(n):
    """
    Calcula el factorial de un n√∫mero.
    
    # Agreg√° ejemplos aqu√≠
    Examples:
        >>> factorial(5)
        120
    """
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
````
