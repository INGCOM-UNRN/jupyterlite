## 6. Operadores L√≥gicos

### Ejercicio Resuelto: Combinar condiciones

**Problema:** Usar and, or, not.

**Soluci√≥n paso a paso:**

````python
# Paso 1: Definir condiciones
edad = 20
tiene_dni = True
es_argentino = True
esta_inhabilitado = False

# Paso 2: AND - todas verdaderas
cumple_edad = edad >= 18
puede_votar = cumple_edad and tiene_dni and es_argentino
print(f"¬øPuede votar? {puede_votar}")

# Paso 3: OR - al menos una verdadera
es_finde = True
es_feriado = False
puede_descansar = es_finde or es_feriado
print(f"¬øPuede descansar? {puede_descansar}")

# Paso 4: NOT - invertir
esta_lloviendo = False
buen_tiempo = not esta_lloviendo
print(f"¬øBuen tiempo? {buen_tiempo}")

# Paso 5: Combinaci√≥n compleja
acceso_permitido = puede_votar and (not esta_inhabilitado)
print(f"¬øAcceso permitido? {acceso_permitido}")
````
