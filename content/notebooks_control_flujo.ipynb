## üîÄ Parte 1: Estructuras Condicionales

### Conceptos Clave
- **`if`**: Ejecuta c√≥digo si una condici√≥n es verdadera
- **`elif`**: Verifica condiciones alternativas
- **`else`**: Ejecuta c√≥digo si ninguna condici√≥n anterior fue verdadera
- La **indentaci√≥n** (sangr√≠a) define qu√© c√≥digo pertenece a cada bloque

### ‚úÖ Ejercicio Resuelto 1.1: Saludo seg√∫n la Hora

**Objetivo:** Usar `if/elif/else` para tomar decisiones.

````python
# Ejercicio Resuelto: Saludo personalizado seg√∫n la hora

hora = 14

if hora < 12:
    print("Buenos d√≠as ‚òÄÔ∏è")
elif hora < 20:
    print("Buenas tardes üå§Ô∏è")
else:
    print("Buenas noches üåô")

print(f"(Son las {hora} horas)")
````

## üîÅ Parte 2: Lazos con `while`

### Conceptos Clave
- **`while`**: Repite c√≥digo mientras una condici√≥n sea verdadera
- Cuidado con los **lazos infinitos**: asegurate de que la condici√≥n eventualmente sea falsa
- Us√° **banderas** (variables booleanas) para controlar la ejecuci√≥n

### ‚úÖ Ejercicio Resuelto 2.1: Contador Simple

**Objetivo:** Entender la estructura b√°sica de `while`.

````python
# Ejercicio Resuelto: Contar de 1 a 10

contador = 1

while contador <= 10:
    print(f"Contador: {contador}")
    contador += 1  # ¬°IMPORTANTE! Modificamos la variable

print("Fin del conteo")
````

### ‚úÖ Ejercicio Resuelto 2.2: Suma Acumulativa

**Objetivo:** Usar un acumulador dentro de un lazo.

````python
# Ejercicio Resuelto: Sumar n√∫meros de 1 a N

n = 100
suma = 0
i = 1

while i <= n:
    suma += i
    i += 1

print(f"La suma de 1 a {n} es: {suma}")

# Verificaci√≥n con la f√≥rmula: n*(n+1)/2
formula = n * (n + 1) // 2
print(f"Verificaci√≥n con f√≥rmula: {formula}")
````

## üîÇ Parte 3: Lazos con `for`

### Conceptos Clave
- **`for`**: Itera sobre una secuencia (lista, rango, string, etc.)
- **`range(n)`**: Genera n√∫meros de 0 a n-1
- **`range(inicio, fin)`**: Genera n√∫meros de inicio a fin-1
- **`range(inicio, fin, paso)`**: Genera n√∫meros con un salto espec√≠fico

### ‚úÖ Ejercicio Resuelto 3.1: Iterar con `range()`

**Objetivo:** Entender las diferentes formas de usar `range()`.

````python
# Ejercicio Resuelto: Formas de usar range()

# range(n): de 0 a n-1
print("De 0 a 4:")
for i in range(5):
    print(i, end=" ")
print()

# range(inicio, fin): de inicio a fin-1
print("\nDe 5 a 10:")
for i in range(5, 11):
    print(i, end=" ")
print()

# range(inicio, fin, paso): con saltos
print("\nPares de 0 a 10:")
for i in range(0, 11, 2):
    print(i, end=" ")
print()

# Cuenta regresiva
print("\nCuenta regresiva:")
for i in range(10, 0, -1):
    print(i, end=" ")
print("¬°Despegue! üöÄ")
````

### ‚úÖ Ejercicio Resuelto 3.2: Suma de Pares

**Objetivo:** Usar `for` con un acumulador.

````python
# Ejercicio Resuelto: Sumar n√∫meros pares de 1 a N

n = 20
suma = 0

for i in range(2, n + 1, 2):  # Empieza en 2, salta de 2 en 2
    suma += i

print(f"La suma de los pares de 1 a {n} es: {suma}")
````

### ‚úÖ Ejercicio Resuelto 3.3: Factorial

**Objetivo:** Calcular el factorial usando `for`.

````python
# Ejercicio Resuelto: Calcular factorial
# n! = 1 √ó 2 √ó 3 √ó ... √ó n

n = 5
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print(f"{n}! = {factorial}")
````

## ‚ö° Parte 4: Control de Lazos (`break` y `continue`)

### Conceptos Clave
- **`break`**: Sale inmediatamente del lazo
- **`continue`**: Salta a la siguiente iteraci√≥n del lazo
- √ötiles para optimizar y controlar el flujo

### ‚úÖ Ejercicio Resuelto 4.1: Uso de `break`

**Objetivo:** Salir de un lazo cuando se cumple una condici√≥n.

````python
# Ejercicio Resuelto: Buscar un n√∫mero en una secuencia

objetivo = 7
encontrado = False

for i in range(1, 101):
    if i == objetivo:
        print(f"¬°Encontr√© el {objetivo} en la posici√≥n {i}!")
        encontrado = True
        break  # Salimos del lazo, no necesitamos seguir buscando

if not encontrado:
    print(f"No se encontr√≥ el n√∫mero {objetivo}")
````

### ‚úÖ Ejercicio Resuelto 4.2: Uso de `continue`

**Objetivo:** Saltar elementos que no interesan.

````python
# Ejercicio Resuelto: Mostrar solo los impares

print("N√∫meros impares del 1 al 20:")
for i in range(1, 21):
    if i % 2 == 0:  # Si es par
        continue  # Saltamos a la siguiente iteraci√≥n
    print(i, end=" ")
````

## üîÑ Parte 5: Lazos Anidados

### Conceptos Clave
- Un lazo **dentro** de otro lazo
- El lazo interno se ejecuta completamente por cada iteraci√≥n del lazo externo
- √ötil para trabajar con estructuras bidimensionales (tablas, matrices, etc.)

### ‚úÖ Ejercicio Resuelto 5.1: Tabla de Multiplicar Completa

**Objetivo:** Usar lazos anidados para generar m√∫ltiples tablas.

````python
# Ejercicio Resuelto: Tablas de multiplicar del 1 al 5

for tabla in range(1, 6):  # Lazo externo: qu√© tabla
    print(f"\nTabla del {tabla}:")
    for i in range(1, 11):  # Lazo interno: multiplicaciones
        resultado = tabla * i
        print(f"{tabla} x {i} = {resultado}")
````

### ‚úÖ Ejercicio Resuelto 5.2: Patr√≥n de Asteriscos
