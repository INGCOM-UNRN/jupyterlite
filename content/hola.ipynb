## ¿Qué es JupyterLab?

JupyterLab es un entorno de trabajo pensado para crear documentos que mezclan texto (como este) con código, como el que está a continuación.

Funciona como un "cuaderno digital" donde podés combinar código, texto explicativo y resultados en un mismo lugar.

En este caso, estamos utilizando una versión JupyterLite. Esto significa que el entorno se ejecuta enteramente en tu navegador web, sin depender de un servidor externo.

    ✅ <b>Tip:</b> Si querés conservar o compartir los cuadernos, los tenés que descargar.= a tu computadora

Dado que esta versión se ejecuta localmente en tu navegador, nosotros no tenemos acceso a tus datos ni a los cambios que realices.
