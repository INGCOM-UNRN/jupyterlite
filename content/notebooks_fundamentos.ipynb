## üî¢ Parte 2: Operadores Aritm√©ticos

### Conceptos Clave
- **Suma:** `+`
- **Resta:** `-`
- **Multiplicaci√≥n:** `*`
- **Divisi√≥n:** `/` (resultado float)
- **Divisi√≥n entera:** `//` (resultado int)

### ‚úÖ Ejercicio Resuelto 2.1: Calculadora B√°sica

**Objetivo:** Realizar operaciones aritm√©ticas b√°sicas.

```python
# Ejercicio Resuelto: Operaciones b√°sicas

a = 15
b = 4

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicaci√≥n: {a} * {b} = {a * b}")
print(f"Divisi√≥n: {a} / {b} = {a / b}")
print(f"Divisi√≥n entera: {a} // {b} = {a // b}")
print(f"M√≥dulo: {a} % {b} = {a % b}")
print(f"Potencia: {a} ** {b} = {a ** b}")
```

## ‚öñÔ∏è Parte 3: Operadores de Comparaci√≥n y L√≥gicos

### Conceptos Clave

**Comparaci√≥n:**
- `==` igual a
- `!=` diferente de
- `<` menor que
- `>` mayor que
- `<=` menor o igual
- `>=` mayor o igual

**L√≥gicos:**
- `and` Y l√≥gico
- `or` O l√≥gico
- `not` negaci√≥n

### ‚úÖ Ejercicio Resuelto 3.1: Comparaciones

**Objetivo:** Practicar operadores de comparaci√≥n.

```python
# Ejercicio Resuelto: Comparaciones

edad = 18
edad_minima_votar = 16
edad_minima_conducir = 18

puede_votar = edad >= edad_minima_votar
puede_conducir = edad >= edad_minima_conducir

print(f"Edad: {edad}")
print(f"¬øPuede votar? {puede_votar}")
print(f"¬øPuede conducir? {puede_conducir}")
```

### ‚úÖ Ejercicio Resuelto 3.2: Operadores L√≥gicos

**Objetivo:** Combinar condiciones con `and`, `or`, `not`.

```python
# Ejercicio Resuelto: L√≥gica booleana

tiene_entrada = True
tiene_dinero = True
es_mayor_edad = False

puede_entrar_cine = tiene_entrada and tiene_dinero
puede_ver_pelicula_adultos = puede_entrar_cine and es_mayor_edad

print(f"¬øPuede entrar al cine? {puede_entrar_cine}")
print(f"¬øPuede ver pel√≠cula para mayores? {puede_ver_pelicula_adultos}")
print(f"¬øEs menor de edad? {not es_mayor_edad}")
```

## üîÑ Parte 5: Conversi√≥n de Tipos (Casting)

### Conceptos Clave
- `int()` convierte a entero
- `float()` convierte a decimal
- `str()` convierte a texto
- `bool()` convierte a booleano
- ¬°Cuidado con las conversiones inv√°lidas!

### ‚úÖ Ejercicio Resuelto 5.1: Conversiones B√°sicas

**Objetivo:** Practicar conversi√≥n entre tipos.

```python
# Ejercicio Resuelto: Conversiones

# String a int
texto_numero = "42"
numero = int(texto_numero)
print(f"'{texto_numero}' (str) -> {numero} (int)")

# Float a int (trunca decimales)
decimal = 3.99
entero = int(decimal)
print(f"{decimal} (float) -> {entero} (int)")

# Int a float
entero = 5
decimal = float(entero)
print(f"{entero} (int) -> {decimal} (float)")

# N√∫mero a string
numero = 100
texto = str(numero)
print(f"{numero} (int) -> '{texto}' (str)")
```
