#### Soluci√≥n Ejercicio 4.1

````python
# SOLUCI√ìN
def validar_email(email):
    """Valida si un string tiene formato de email b√°sico.
    
    Criterios de validaci√≥n:
    - Debe contener exactamente un @
    - Debe tener texto antes y despu√©s del @
    - Debe tener un punto despu√©s del @
    
    Args:
        email (str): String a validar
        
    Returns:
        bool: True si el formato es v√°lido, False en caso contrario
        
    Examples:
        >>> validar_email("usuario@ejemplo.com")
        True
        >>> validar_email("invalido@")
        False
    """
    # Verificar que tiene exactamente un @
    if email.count('@') != 1:
        return False
    
    # Dividir en partes
    partes = email.split('@')
    parte_usuario = partes[0]
    parte_dominio = partes[1]
    
    # Verificar que ambas partes no est√©n vac√≠as
    if len(parte_usuario) == 0 or len(parte_dominio) == 0:
        return False
    
    # Verificar que el dominio tenga un punto
    if '.' not in parte_dominio:
        return False
    
    # Todo v√°lido
    return True

# Pruebas
emails_prueba = [
    "usuario@ejemplo.com",
    "invalido@",
    "@ejemplo.com",
    "usuario@@ejemplo.com",
    "usuario@ejemplo",
    "nombre.apellido@empresa.com.ar",
]

print("Validaci√≥n de emails:\n")
for email in emails_prueba:
    resultado = validar_email(email)
    simbolo = "‚úì" if resultado else "‚úó"
    print(f"{simbolo} {email:35s} -> {resultado}")
````

### Ejercicio 4.4: Funci√≥n Recursiva - Factorial

**Descripci√≥n**: Implementa el c√°lculo del factorial de forma recursiva.

**Especificaciones**:
- Entrada: N√∫mero entero n >= 0
- Salida: Factorial de n
- Recordatorio: n! = n √ó (n-1) √ó (n-2) √ó ... √ó 1
- Caso base: 0! = 1, 1! = 1
- Caso recursivo: n! = n √ó (n-1)!

**Pista**: La funci√≥n debe llamarse a s√≠ misma.

````python
# TU C√ìDIGO AQU√ç
def factorial(n):
    """Calcula el factorial de n de forma recursiva.
    
    Args:
        n (int): N√∫mero entero no negativo
        
    Returns:
        int: Factorial de n
    """
    # Tu c√≥digo aqu√≠
    pass

# Pruebas
for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

#### Soluci√≥n Ejercicio 4.4

````python
# SOLUCI√ìN
def factorial(n):
    """Calcula el factorial de n de forma recursiva.
    
    El factorial de n (n!) es el producto de todos los enteros
    positivos desde 1 hasta n.
    
    Args:
        n (int): N√∫mero entero no negativo
        
    Returns:
        int: Factorial de n (n!)
        
    Examples:
        >>> factorial(5)
        120
        >>> factorial(0)
        1
    """
    # Caso base: 0! = 1 y 1! = 1
    if n <= 1:
        return 1
    
    # Caso recursivo: n! = n √ó (n-1)!
    return n * factorial(n - 1)

# Pruebas
print("C√°lculo de factoriales:\n")
for i in range(8):
    resultado = factorial(i)
    print(f"{i}! = {resultado}")

# Verificaci√≥n manual
print("\nVerificaci√≥n:")
print(f"5! = 5 √ó 4 √ó 3 √ó 2 √ó 1 = {5*4*3*2*1}")
print(f"factorial(5) = {factorial(5)}")
````

#### Soluci√≥n Ejercicio 4.5

````python
# SOLUCI√ìN
def sumar_todos(*args):
    """Suma cualquier cantidad de n√∫meros.
    
    Args:
        *args: N√∫meros a sumar (int o float)
        
    Returns:
        float: Suma de todos los argumentos, 0 si no hay argumentos
        
    Examples:
        >>> sumar_todos(1, 2, 3)
        6
        >>> sumar_todos()
        0
        >>> sumar_todos(5.5, 3.2, 1.3)
        10.0
    """
    # Si no hay argumentos, retornar 0
    if len(args) == 0:
        return 0
    
    # Validar que todos sean n√∫meros
    for arg in args:
        if not isinstance(arg, (int, float)):
            print(f"Advertencia: {arg} no es un n√∫mero, se ignora")
    
    # Sumar todos los argumentos num√©ricos
    suma = 0
    for numero in args:
        if isinstance(numero, (int, float)):
            suma += numero
    
    return suma

# Pruebas
print("Suma con n√∫mero variable de argumentos:\n")
print(f"sumar_todos(1, 2, 3) = {sumar_todos(1, 2, 3)}")
print(f"sumar_todos(10, 20, 30, 40, 50) = {sumar_todos(10, 20, 30, 40, 50)}")
print(f"sumar_todos() = {sumar_todos()}")
print(f"sumar_todos(5.5, 3.2, 1.8) = {sumar_todos(5.5, 3.2, 1.8)}")
print(f"\nCon un solo argumento:")
print(f"sumar_todos(42) = {sumar_todos(42)}")
````

#### Soluci√≥n Ejercicio Integrador

````python
# SOLUCI√ìN COMPLETA

# Estructura de datos global
estudiantes = {}

def agregar_estudiante(nombre):
    """Agrega un nuevo estudiante al sistema.
    
    Args:
        nombre (str): Nombre del estudiante
        
    Returns:
        bool: True si se agreg√≥, False si ya exist√≠a
    """
    if nombre in estudiantes:
        print(f"‚ö†Ô∏è  El estudiante '{nombre}' ya existe")
        return False
    
    estudiantes[nombre] = []
    print(f"‚úì Estudiante '{nombre}' agregado")
    return True

def agregar_nota(nombre, nota):
    """Agrega una nota a un estudiante existente.
    
    Args:
        nombre (str): Nombre del estudiante
        nota (float): Nota a agregar (0-100)
        
    Returns:
        bool: True si se agreg√≥, False si hubo error
    """
    # Validar que el estudiante existe
    if nombre not in estudiantes:
        print(f"‚úó El estudiante '{nombre}' no existe")
        return False
    
    # Validar nota
    if nota < 0 or nota > 100:
        print(f"‚úó Nota inv√°lida: {nota}. Debe estar entre 0 y 100")
        return False
    
    estudiantes[nombre].append(nota)
    print(f"‚úì Nota {nota} agregada a {nombre}")
    return True

def calcular_promedio(nombre):
    """Calcula el promedio de notas de un estudiante.
    
    Args:
        nombre (str): Nombre del estudiante
        
    Returns:
        float: Promedio de notas, None si no existe o no tiene notas
    """
    if nombre not in estudiantes:
        return None
    
    notas = estudiantes[nombre]
    if len(notas) == 0:
        return None
    
    return sum(notas) / len(notas)

def listar_estudiantes():
    """Lista todos los estudiantes con sus estad√≠sticas."""
    if len(estudiantes) == 0:
        print("No hay estudiantes registrados")
        return
    
    print("\n" + "="*60)
    print("LISTADO DE ESTUDIANTES")
    print("="*60)
    
    for nombre in sorted(estudiantes.keys()):
        notas = estudiantes[nombre]
        promedio = calcular_promedio(nombre)
        
        print(f"\n{nombre}:")
        print(f"  Notas: {notas}")
        if promedio is not None:
            print(f"  Promedio: {promedio:.2f}")
        else:
            print(f"  Promedio: Sin notas")
    
    print("\n" + "="*60)

def mejor_estudiante():
    """Encuentra el estudiante con el mejor promedio.
    
    Returns:
        tuple: (nombre, promedio) del mejor estudiante
               None si no hay estudiantes con notas
    """
    if len(estudiantes) == 0:
        return None
    
    mejor_nombre = None
    mejor_promedio = -1
    
    for nombre in estudiantes:
        promedio = calcular_promedio(nombre)
        if promedio is not None and promedio > mejor_promedio:
            mejor_promedio = promedio
            mejor_nombre = nombre
    
    if mejor_nombre is None:
        return None
    
    return (mejor_nombre, mejor_promedio)

# PRUEBAS DEL SISTEMA
print("=" * 60)
print("SISTEMA DE GESTI√ìN DE ESTUDIANTES")
print("=" * 60)

# Agregar estudiantes
print("\n1. Agregando estudiantes:")
agregar_estudiante("Ana")
agregar_estudiante("Carlos")
agregar_estudiante("Elena")
agregar_estudiante("Ana")  # Duplicado

# Agregar notas
print("\n2. Agregando notas:")
agregar_nota("Ana", 85)
agregar_nota("Ana", 90)
agregar_nota("Ana", 88)
agregar_nota("Carlos", 78)
agregar_nota("Carlos", 82)
agregar_nota("Elena", 95)
agregar_nota("Elena", 92)
agregar_nota("Elena", 97)
agregar_nota("Bruno", 80)  # Estudiante no existe

# Listar todos
print("\n3. Listado completo:")
listar_estudiantes()

# Mejor estudiante
print("\n4. Mejor estudiante:")
mejor = mejor_estudiante()
if mejor is not None:
    nombre, promedio = mejor
    print(f"üèÜ {nombre} con promedio de {promedio:.2f}")

# Calcular promedio espec√≠fico
print("\n5. Consultas individuales:")
for nombre in ["Ana", "Carlos", "Elena"]:
    prom = calcular_promedio(nombre)
    print(f"  Promedio de {nombre}: {prom:.2f}")
````
