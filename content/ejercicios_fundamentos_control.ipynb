### Ejercicio 2.3: Validaci√≥n de Contrase√±a con Bucle

**Descripci√≥n**: Simula un sistema de login que da 3 intentos para ingresar la contrase√±a correcta.

**Especificaciones**:
- Contrase√±a correcta: "python123"
- M√°ximo 3 intentos
- Usa `while` con una bandera de control
- Muestra mensajes apropiados

**Nota**: Como estamos en Jupyter, simularemos los intentos con una lista predefinida en lugar de `input()`.

````python
# TU C√ìDIGO AQU√ç
# Simulaci√≥n de intentos (en un programa real usar√≠as input())
intentos_simulados = ["contrase√±a", "python", "python123"]  # Prueba con diferentes secuencias

CONTRASENA_CORRECTA = "python123"
MAX_INTENTOS = 3

# Tu c√≥digo aqu√≠

````

#### Soluci√≥n Ejercicio 2.3

````python
# SOLUCI√ìN
intentos_simulados = ["contrase√±a", "python", "python123"]

CONTRASENA_CORRECTA = "python123"
MAX_INTENTOS = 3

intentos = 0
acceso_concedido = False

while intentos < MAX_INTENTOS and not acceso_concedido:
    # Simulamos input con la lista
    contrasena_ingresada = intentos_simulados[intentos]
    print(f"Intento {intentos + 1}: {contrasena_ingresada}")
    
    if contrasena_ingresada == CONTRASENA_CORRECTA:
        acceso_concedido = True
        print("‚úì Acceso concedido")
    else:
        intentos += 1
        if intentos < MAX_INTENTOS:
            print(f"‚úó Contrase√±a incorrecta. Te quedan {MAX_INTENTOS - intentos} intentos")

if not acceso_concedido:
    print("‚úó Acceso denegado. Has agotado tus intentos.")
````

**Descripci√≥n**: Calcula la suma de todos los n√∫meros pares entre 1 y N (inclusive).

**Especificaciones**:
- Entrada: N√∫mero entero N
- Salida: Suma de todos los pares hasta N
- Usa un `for` loop con `range()`
- Usa el operador m√≥dulo `%` para verificar si es par

**Ejemplo**:
```
N = 10
Pares: 2, 4, 6, 8, 10
Suma: 30
```

````python
# TU C√ìDIGO AQU√ç
n = 10

# Calcula la suma de pares

````

#### Soluci√≥n Ejercicio 2.4

````python
# SOLUCI√ìN
n = 10
suma_pares = 0

for numero in range(1, n + 1):
    if numero % 2 == 0:
        suma_pares += numero
        print(f"Par encontrado: {numero}, suma acumulada: {suma_pares}")

print(f"\nSuma total de pares hasta {n}: {suma_pares}")
````

**Especificaciones**:
- Entrada: N√∫mero entero N > 1
- Salida: Mensaje indicando si es primo o no
- Usa un `for` loop para verificar divisibilidad
- Usa una bandera de control

**Recordatorio**: Un n√∫mero es primo si solo es divisible por 1 y por s√≠ mismo.

**Pista**: Solo necesit√°s verificar divisores hasta la ra√≠z cuadrada de N, pero para simplificar, pod√©s verificar hasta N-1.

````python
# TU C√ìDIGO AQU√ç
numero = 17  # Prueba con: 17 (primo), 18 (no primo), 2 (primo), 1 (caso especial)

# Determina si es primo

````

#### Soluci√≥n Ejercicio 2.6

````python
# SOLUCI√ìN
numero = 17

# Casos especiales
if numero < 2:
    es_primo = False
elif numero == 2:
    es_primo = True
else:
    # Verificar si tiene divisores
    es_primo = True  # Asumimos que es primo
    
    for divisor in range(2, numero):
        if numero % divisor == 0:
            es_primo = False  # Encontramos un divisor, no es primo
            print(f"Divisor encontrado: {divisor}")
            # En un programa real, aqu√≠ podr√≠as usar break,
            # pero en este curso preferimos banderas

if es_primo:
    print(f"‚úì {numero} es un n√∫mero primo")
else:
    print(f"‚úó {numero} NO es un n√∫mero primo")
````

### Ejercicio Integrador 1: Adivina el N√∫mero

**Descripci√≥n**: Simula un juego donde el programa "piensa" un n√∫mero y el usuario tiene que adivinarlo.

**Especificaciones**:
- N√∫mero secreto predefinido (simulaci√≥n)
- M√°ximo 5 intentos
- Dar pistas: "demasiado alto" o "demasiado bajo"
- Usa variables de control y condicionales

**Nota**: Simularemos los intentos con una lista.

````python
# TU C√ìDIGO AQU√ç
# Simulaci√≥n de intentos
intentos_usuario = [50, 75, 85, 90, 87]  # Prueba con diferentes secuencias

NUMERO_SECRETO = 87
MAX_INTENTOS = 5

# Tu c√≥digo aqu√≠

````

#### Soluci√≥n Ejercicio Integrador 1

````python
# SOLUCI√ìN
intentos_usuario = [50, 75, 85, 90, 87]

NUMERO_SECRETO = 87
MAX_INTENTOS = 5

intentos = 0
adivinado = False

print(f"¬°Bienvenido al juego! Ten√©s {MAX_INTENTOS} intentos para adivinar el n√∫mero entre 1 y 100.\n")

while intentos < MAX_INTENTOS and not adivinado:
    numero_ingresado = intentos_usuario[intentos]
    intentos += 1
    
    print(f"Intento {intentos}: {numero_ingresado}")
    
    if numero_ingresado == NUMERO_SECRETO:
        adivinado = True
        print(f"\n¬°üéâ Felicitaciones! Adivinaste el n√∫mero en {intentos} intentos.")
    elif numero_ingresado < NUMERO_SECRETO:
        print("‚Üí El n√∫mero es m√°s alto\n")
    else:
        print("‚Üí El n√∫mero es m√°s bajo\n")

if not adivinado:
    print(f"\nüòû Se acabaron los intentos. El n√∫mero era {NUMERO_SECRETO}")
````

### Ejercicio Integrador 2: Calculadora de Promedio con Validaci√≥n

**Descripci√≥n**: Calcula el promedio de N notas, validando que est√©n en el rango correcto.

**Especificaciones**:
- Notas entre 0 y 100
- Rechazar notas inv√°lidas y pedirlas nuevamente
- Calcular y mostrar el promedio al final
- Mostrar la clasificaci√≥n seg√∫n el promedio

**Integra**: Variables, tipos, operadores, condicionales, loops, validaci√≥n

````python
# TU C√ìDIGO AQU√ç
# Simulaci√≥n de notas ingresadas
notas_simuladas = [85, 90, -5, 105, 78, 92]  # Incluye valores inv√°lidos

CANTIDAD_NOTAS = 3

# Tu c√≥digo aqu√≠

````

#### Soluci√≥n Ejercicio Integrador 2

````python
# SOLUCI√ìN
notas_simuladas = [85, 90, -5, 105, 78, 92]
CANTIDAD_NOTAS = 3

notas_validas = []
indice_simulacion = 0

print(f"Ingresa {CANTIDAD_NOTAS} notas (entre 0 y 100):\n")

while len(notas_validas) < CANTIDAD_NOTAS:
    # Simulamos input
    nota = notas_simuladas[indice_simulacion]
    indice_simulacion += 1
    
    print(f"Nota {len(notas_validas) + 1}: {nota}")
    
    # Validaci√≥n
    if nota < 0 or nota > 100:
        print("‚úó Nota inv√°lida. Debe estar entre 0 y 100.\n")
    else:
        notas_validas.append(nota)
        print(f"‚úì Nota aceptada\n")

# Calcular promedio
suma = 0
for nota in notas_validas:
    suma += nota

promedio = suma / CANTIDAD_NOTAS

# Clasificar
if promedio >= 90:
    clasificacion = "Excelente"
elif promedio >= 80:
    clasificacion = "Muy Bueno"
elif promedio >= 70:
    clasificacion = "Bueno"
elif promedio >= 60:
    clasificacion = "Aprobado"
else:
    clasificacion = "Reprobado"

print(f"\n{'='*40}")
print(f"Notas: {notas_validas}")
print(f"Promedio: {promedio:.2f}")
print(f"Clasificaci√≥n: {clasificacion}")
print(f"{'='*40}")
````
